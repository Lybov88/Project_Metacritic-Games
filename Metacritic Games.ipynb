```python
#создадим новый ДФ, где столбец metascore будет типа int
df_met_int = df.query("metascore != 'tbd'").astype({'metascore': int})
```

```python
#посмотрим metascore по платформам
df_met_int.groupby('platform', as_index=False) \
        .agg({'metascore': 'sum'}) \
        .sort_values('metascore', ascending=False) \
        .head(20)
```
